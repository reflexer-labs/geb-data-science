```python
import matplotlib.pyplot as plt
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
from matplotlib import cm
import numpy as np

from mlxtend.plotting import heatmap
```

```python
discount = np.round(np.linspace(0.85, 0.99, 15).reshape(-1, 1),2)  # len = 11
penalty = np.round(np.linspace(1, 1.2, 21).reshape(1, -1), 2)

less_than_penalty_raised = 1/discount * penalty
discount.shape, penalty.shape, less_than_penalty_raised .shape
```

```python
penalty
```

```python
heatmap(less_than_penalty_raised , column_names=list(map(str, penalty.flatten())),
        row_names=list(map(str, discount.flatten())),
        figsize=(20, 10),
        cell_values=True)
        #cmap='Greens')
plt.title('Minimum liquidation ratio at which full liquidation penalty is raised', fontsize=24)
plt.xlabel('liquidation penalty', fontsize=20)
plt.ylabel('collateral discount', fontsize=20)
plt.show()
```

For each pair of discount/penalty values, the above heatmap shows the actual liquidation ratio at which the system starts 
to receive less surplus per liquidation.

Example:
If `discount=0.96` and `penalty=1.15`, then all liquidations that occur at ratio=`1.20` will receive less than the full penalty amount.

```python
discount = np.round(np.linspace(0.8, 0.96, 15).reshape(-1, 1),2)  # len = 11
bad_debt_ratio = 1/discount

plt.figure(figsize=(8, 5))
plt.plot(discount, bad_debt_ratio)
plt.xlabel('Collateral discount', fontsize=20)
plt.ylabel('Actual Liquidation Ratio', fontsize=20)
plt.title('Minimum Actual Liquidation Ratio w/o Bad Debt', fontsize=20)
plt.show()
```

The above charts shows theliquidation ratio at which bad debt will be created for different collateral discounts.
